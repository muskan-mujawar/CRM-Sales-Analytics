```python
import warnings
import datetime
warnings.filterwarnings('ignore')

from sqlalchemy import  create_engine

import pandas as pd

engine = create_engine("mysql+pymysql://root:root@127.0.0.1:3306/crm_sales")

# {root}:{password}@host:port/{database name}

#Establish a connection

conn = engine.connect()


#Write DataFrame to a SQL table

df1.to_sql('sales_team', engine, index=False, if_exists='replace')
df2.to_sql('sales_pipeline', engine, index=False, if_exists='replace')
df3.to_sql('products', engine, index=False, if_exists='replace')
df4.to_sql('accounts', engine, index=False, if_exists='replace')

#Close Connection
conn.close()
print("done")
```
